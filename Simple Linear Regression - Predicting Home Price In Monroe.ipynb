```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('data/homeprices.csv')
```

```python
df.head()
```

```python
df.tail()
```

```python
plt.xlabel('Area(Sq.ft)')
plt.ylabel('Price ($)')
plt.scatter(df.area, df.price, color='red', marker="*")
m, b = np.polyfit(df.area, df.price, 1)
plt.plot(df.area, m*df.area + b)
```

```python
area = df.drop('price',axis='columns')
price = df.price
```

```python
reg = linear_model.LinearRegression()
```

```python
reg.fit(area, price)
```

```python
new_areas = [[35000],[38000],[49000]]
reg.predict(new_areas)
```

```python
area_data = pd.read_csv('data/areas.csv')
```

```python
area_data['price'] = reg.predict(area_data)
```

```python
area_data.to_csv('data/houseprice_output.csv', index=False)
```
