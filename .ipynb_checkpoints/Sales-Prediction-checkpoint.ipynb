# Sales Prediction

The aim is to build a model which predicts sales based on the money spent on different platforms such as TV, radio, and newspaper for marketing.

# Model Building

Linear Regression is a useful tool for predicting a quantitative response.

Salesn prediction using multiple linear Regression:

**2. Multiple Linear Regression**

Multiple linear regression has one y and two or more x variables. It is an extension of Simple Linear regression as it takes more than one predictor variable to predict the response variable.

Multiple Linear Regression is one of the important regression algorithms which models the linear relationship between a single dependent continuous variable and more than one independent variable.

Assumptions for Multiple Linear Regression:
    1. A linear relationship should exist between the Target and predictor variables.
    2. The regression residuals must be normally distributed.
    3. MLR assumes little or no multicollinearity (correlation between the independent variable) in data.
    
**Formula:** Y = 0 + 1X1 + 2X2 + 3X3 + ... + nXn + e

    Y = Dependent variable / Target variable
    0 = Intercept of the regression line 
    1, 2,..n = Slope of the regression lime which tells whether the line is increasing or decreasing
    X1, X2,..Xn = Independent variables / Predictor variables
    e = Error
    
**Equation:** Sales = 0 + (1 * TV) + (2 * Radio) + (3 * Newspaper)

```python
#Setting the value for X and Y
X = df[['TV', 'Radio', 'Newspaper']]
Y = df['Sales']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=100)
```

```python
regression = LinearRegression()

# Train the model using the training sets
regression.fit(X_train, Y_train)

# Make predictions using the testing set
score_y_predict = regression.predict(X_test)
```

```python
#Actual value and the predicted value
mlr_compare = pd.DataFrame({'Actual value': Y_test, 'Predicted value': score_y_predict})
mlr_compare
```

```python
df1 = mlr_compare.head(25)
df1.plot(kind='bar',figsize=(16,10))
plt.grid(which='major', linestyle='-', linewidth='0.2', color='green')
plt.grid(which='minor', linestyle=':', linewidth='0.2', color='black')
plt.show()
```

```python
#Predicting the Test and Train set result 
regression.coef_
```

```python
# The coefficients
print("Coefficients: \n", regression.coef_)
# The mean squared error
print("Mean squared error: %.2f" % mean_squared_error(Y_test, score_y_predict))
# The coefficient of determination: 1 is perfect prediction
print("Coefficient of determination: %.2f" % r2_score(Y_test, score_y_predict))

print('Mean Absolute Error:', mean_absolute_error(Y_test, score_y_predict))  
print('Mean Squared Error:', mean_squared_error(Y_test, score_y_predict))  
print('Root Mean Squared Error:', np.sqrt(mean_squared_error(Y_test, score_y_predict)))
print('R squared value of the model: {:.2f}'.format(regression.score(X,Y)*100))
```

**Conclusion:** 90.21% of the data fit the multiple regression model
