```python
import pandas as pd
import numpy as np
from sklearn import linear_model 
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv("data/homeprices_categorical_variables_and_one_hot_encoding.csv")
```

```python
df
```

```python
dummies = pd.get_dummies(df.town)
```

```python
merged = pd.concat([df,dummies], axis='columns')
```

```python
merged
```

```python
final = merged.drop(['town','west windsor'], axis='columns')
```

```python
final
```

```python
model = linear_model.LinearRegression()
```

```python
X = final.drop('price', axis='columns')
```

```python
y = df.price
```

```python
model.fit(X,y)
```

### Robinsville

```python
model.predict([[2800,0,1]])
```

### West windsor

```python
model.predict([[2800,0,0]])
```

### how accurate our model is ?

```python
model.score(X,y)
```

## Using `Sklearn OneHotEncoder`

```python
df
```

```python
le = LabelEncoder()
```

```python
dfle = df
dfle.town = le.fit_transform(dfle.town)
dfle
```

```python
X = dfle[['town','area']].values
```

```python
X
```

```python
y = dfle.price.values
y
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
ct = ColumnTransformer([('town', OneHotEncoder(), [0])], remainder = 'passthrough')
```

```python
X = ct.fit_transform(X)
X
```

### To avoid dummy variable trap <a href="https://en.wikipedia.org/wiki/Dummy_variable_(statistics)">Read Dummy variable</a>

```python
X = X[:,1:]
X
```

```python
model.fit(X,y)
```

```python
model.predict([[0,1,3400]]) # 3400 sqr ft home in west windsor
```

```python
model.predict([[1,0,2800]]) # 2800 sqr ft home in robbinsville
```
