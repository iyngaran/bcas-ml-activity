```python
import pandas as pd
import numpy as np
from sklearn import linear_model 
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("data/canada_per_capita_income.csv")
```

```python
df.head()
```

```python
df.tail()
```

```python
plt.xlabel("Year")
plt.ylabel("Per capita income (US$)")
plt.scatter(df['year'], df['per capita income (US$)'], color='red', marker="*")

m, b = np.polyfit(df['year'], df['per capita income (US$)'], 1)
plt.plot(df['year'], m*df['year'] + b)
```

```python
reg = linear_model.LinearRegression()
x_year = df.drop("per capita income (US$)", axis='columns')
y_icome = df['per capita income (US$)']
reg.fit(x_year, y_icome)
```

```python
reg.predict([[2018]])
```

# Save and Reuse the model with pickle

```python
import pickle
```

```python
with open('models/canada_per_capita_income', 'wb') as f:
    pickle.dump(reg,f)
```

```python
with open('models/canada_per_capita_income','rb') as f:
    c_i_model = pickle.load(f) 
```

```python
c_i_model.predict([[2018]])
```

# Save and Reuse the model with joblib

```python
import joblib
```

```python
joblib.dump(reg, "models/joblib_canada_per_capita_income")
```

```python
mj = joblib.load("models/joblib_canada_per_capita_income")
```
